```python
import re
import pandas as pd
import nltk
import contractions
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    if not isinstance(text, str):
        return ""  # or you can return np.nan if preferred
    words = re.findall(r'\b\w+\b', text.lower())
    filtered = [word for word in words if word not in stop_words]
    return " ".join(filtered)
```

```python
# Load dataset
df1 = pd.read_csv("final_cleaned_dataset.csv")  # must contain 

df1['no_stopwords_post_content'] = df1['post_content_llm'].apply(remove_stopwords)
df1['no_stopwords_comments_content'] = df1['comments_content_llm'].apply(remove_stopwords)
df1['no_stopwords_author_comment'] = df1['authors_replies_llm'].apply(remove_stopwords)

# Save updated file
df1.to_csv("no_stopword_data.csv", index=False)
```
